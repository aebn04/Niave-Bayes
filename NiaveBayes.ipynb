```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("drive/MyDrive/ML/Praktikum1/AkmalNajib_202010225167_NaiveBayes.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:, :-1]
x = x.apply(LabelEncoder().fit_transform)
x
```

```python
y = dataset.iloc[:, 4]
y
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.5, random_state=322)
```

```python
from sklearn.naive_bayes import GaussianNB
#Mengaktifkan/Memanggil/Membuat fungsi klasifikasi naive bayes
modelnb = GaussianNB()
#Memasukan data Train pada fungsi klasifikasi naive bayes
nbtrain = modelnb.fit(xTrain, yTrain)
```

```python
#Menentukan hasil prediksi dari xTest
yPrediction = nbtrain.predict(xTest)
yPrediction
```

```python
np.array(yTest)
```

```python
nbtrain.predict_proba(xTest)
```

```python
y_actual = pd.Series(np.array(yTest), name = "actual")
y_pred = pd.Series(yPrediction, name = "prediction")
df_confusion = pd.crosstab(y_actual, y_pred)
df_confusion
```

```python
from sklearn.metrics import classification_report
print(classification_report(yTest, yPrediction))
```
